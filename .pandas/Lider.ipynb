```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Quiz Cálculo 1- Gobierno 22 de Octubre(1-18).csv', encoding='utf8')
```

```python
df["n"] = df["Número de documento"].astype(str).apply(
    lambda x: sum(int(i) for i in x))
```

```python
def punto1(n):
    n1=int(n[4]) 
    c105=(n1*105**2)+(2*105)+23
    c100=(n1*100**2)+(2*100)+23
    fc1=(c105-c100)/(105-100)
    return fc1
```

```python
df["Punto 1"]=df.apply(punto1, axis=1)
```

```python
def calificación1(n):
    n1=int(n[5])
    n2=int(n[1])
    if n1==n2:
        return 5
    else:
        return 1
```

```python
df["Calificación P1"]=df.apply(calificación1, axis=1)
```

```python
df.to_csv('Quiz Cálculo 1- Gobierno 22 de Octubre(1-18).csv', index=False)
```

```python
def cf(n):
    nota1=int(n[6]) 
    nota2=int(n[8]) 
    nota3=int(n[10])
    notafinal=(nota1+nota2+nota3)/3 
    return round(notafinal, 2)
```

```python
df["Calificación Final"]=df.apply(cf, axis=1)
```

```python
df.to_csv('Quiz Cálculo 1- Gobierno 22 de Octubre(1-18).csv', index=False)
```

```python
df = df.rename(columns={"Calificación P1": "CP1"})
df = df.rename(columns={"Calificación Final": "NF"})
```

```python
def promedio(na, nc):
    try:
        df = pd.read_csv(na)
        
        if nc not in df.columns:
            return "No existe la columna"

        return df[nc].mean()

    except Exception as e:
        return f"Error al procesar la columna"
```

```python
promedio('Quiz Cálculo 1- Gobierno 22 de Octubre(1-18).csv', 'CP1')
```

```python
def percentil(na, nc):
    try:
        df = pd.read_csv(na)

        if nc not in df.columns:
            return "Error al procesar la columna"

        p25 = df[nc].quantile(0.25)
        p50 = df[nc].quantile(0.50)
        p75 = df[nc].quantile(0.75)

        return p25, p50, p75

    except Exception as e:
        return f"Error al procesar la columna"
```

```python
percentil("Quiz Cálculo 1- Gobierno 22 de Octubre(1-18).csv", "CP1")
```
